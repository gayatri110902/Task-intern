```python
# Importing necessary libraries
import yfinance as yf
import pandas as pd

class StockPortfolio:
    def __init__(self):
        # Initialize an empty portfolio DataFrame
        self.portfolio = pd.DataFrame(columns=['Stock', 'Shares', 'Purchase Price'])

    def add_stock(self, ticker, shares, purchase_price):
        """Add a stock to the portfolio"""
        try:
            stock_data = yf.Ticker(ticker)
            current_price = stock_data.history(period="1d")['Close'][0]  # Get the latest closing price
            new_row = pd.DataFrame([[ticker, shares, purchase_price]], columns=['Stock', 'Shares', 'Purchase Price'])
            self.portfolio = pd.concat([self.portfolio, new_row], ignore_index=True)
            print(f"Added {shares} shares of {ticker} at ${purchase_price} per share.")
        except Exception as e:
            print(f"Error adding stock {ticker}: {e}")

    def remove_stock(self, ticker):
        """Remove a stock from the portfolio"""
        if ticker in self.portfolio['Stock'].values:
            self.portfolio = self.portfolio[self.portfolio['Stock'] != ticker]
            print(f"Removed {ticker} from portfolio.")
        else:
            print(f"{ticker} not found in the portfolio.")

    def track_portfolio(self):
        """Track portfolio value and performance"""
        # Get the latest price for each stock in the portfolio
        self.portfolio['Current Price'] = self.portfolio['Stock'].apply(self.get_current_price)
        self.portfolio['Value'] = self.portfolio['Shares'] * self.portfolio['Current Price']
        self.portfolio['Profit/Loss'] = (self.portfolio['Current Price'] - self.portfolio['Purchase Price']) * self.portfolio['Shares']
        
        total_value = self.portfolio['Value'].sum()
        total_profit_loss = self.portfolio['Profit/Loss'].sum()

        print("\nPortfolio Summary:")
        print(self.portfolio[['Stock', 'Shares', 'Purchase Price', 'Current Price', 'Value', 'Profit/Loss']])
        print(f"\nTotal Portfolio Value: ${total_value:.2f}")
        print(f"Total Profit/Loss: ${total_profit_loss:.2f}")

    def get_current_price(self, ticker):
        """Fetch the current stock price"""
        try:
            stock_data = yf.Ticker(ticker)
            current_price = stock_data.history(period="1d")['Close'][0]
            return current_price
        except Exception as e:
            print(f"Error fetching price for {ticker}: {e}")
            return 0

# Main program to interact with the portfolio
if __name__ == "__main__":
    portfolio = StockPortfolio()

    # Adding stocks
    portfolio.add_stock("AAPL", 10, 150)  # 10 shares of Apple bought at $150 each
    portfolio.add_stock("GOOG", 5, 2800)  # 5 shares of Google bought at $2800 each

    # Tracking portfolio performance
    portfolio.track_portfolio()

    # Removing a stock
    portfolio.remove_stock("AAPL")

    # Tracking portfolio after removal
    portfolio.track_portfolio()
```
